```python
from diffusers import StableDiffusionPipeline
from PIL import Image
import torch

# Stable Diffusion Function
def generate_image_stable_diffusion(prompt, save_path="stable_diffusion_output.png"):
    print(f"Generating image with Stable Diffusion for prompt: {prompt}")

    device = "cuda" if torch.cuda.is_available() else "cpu"
    pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to(device)

    # Generate image
    image = pipe(prompt).images[0]
    image.save(save_path)
    print(f"Stable Diffusion image saved at {save_path}")
    return image

# Main Execution
prompt = "A futuristic cityscape with flying cars at sunset"
image = generate_image_stable_diffusion(prompt)

# Display the image
image.show()
```
